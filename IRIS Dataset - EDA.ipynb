# Exploratory Data Analysis - IRIS Dataset

## Objective: Classify a new flower as belonging to one of the 3 classes given the 4 features.

* The Iris flower data set or Fisher's Iris data set is a multivariate data set introduced by the British statistician and biologist Ronald Fisher in his 1936
* The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimetres. Based on the combination of these four features, Fisher developed a linear discriminant model to distinguish the species from each other.
* 4 Independent Variables- The length and the width of the sepals and petals
* Classification Problem - Classify the flower into SETOSA,VIRGINICA or VERSICOLOR species

* Dataset Description - [https://en.wikipedia.org/wiki/Iris_flower_data_set]
* Dataset Location - [https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv]

```python
# Import required Packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read IRIS Dataset into Iris Dataframe
iris = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
# (Question): Check how many data points and features are there in the Dataset
print(iris.shape)

#(Answer): THe Dataset has 150 rows and 5 columns
```

```python
#(Question): What are the column names in the Dataset?
print(iris.columns)

#(Answer): 5 Columns - 4 input and 1 output
#          Input Variables - Sepal Length, Sepal Width, Petal Length and Petal Width
#          Output Variables - Species(to be predictes)
```

```python
#(Question): How many flowers for each species are present?
iris['species'].value_counts()

#(Answer): Each species has 50 datapoints each. THe IRIS dataset is a balanced dataset.
```

```python
# 2-D Scatter plot with color coding for each species/flower

sns.set_style('whitegrid')
sns.FacetGrid(iris, hue = 'species', size = 4).map(plt.scatter,'sepal_length','sepal_width').add_legend()
plt.show()

#(Observations): 1. Setosa(Blue) species can be easily separated from Versicolor and Virginica by drawing a straight line.
```

```python
# To visualize the relations between the 4 input variables, we create pair-wise scatter plots or Pair-Plot.
plt.close()
sns.set_style('whitegrid')
sns.pairplot(iris, hue = 'species', size = 3)
plt.show()
```

* (Observation1): Petal Length vs Petal Width is the most suitable plot for identifying the three flower types
* (Observation2): Setosa can be directly identified from the plot. However, Versicolor and Virginica have some overlap.

## Univariate Analysis
